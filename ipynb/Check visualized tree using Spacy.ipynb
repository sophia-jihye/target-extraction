```python
import spacy
sample = 'The photo quality is amazing.'

nlp_spacy = spacy.load('en_core_web_sm')
doc = nlp_spacy(sample)
spacy.displacy.serve(doc, style='dep')
```

```python
import spacy
samples = ['Household spending has continued to grow strongly, while growth of business fixed investment has moderated from its rapid pace earlier last year. ', 'Recent indicators point to slower growth of household spending and business fixed investment in the first quarter.']
nlp_spacy = spacy.load('en_core_web_sm')

for sample in samples:
    doc = nlp_spacy(sample)
    spacy.displacy.serve(doc, style='dep')
```
